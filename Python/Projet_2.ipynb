```python
variable_Absences="Absences"
variable_performance="PerformanceScore"
df[[variable_Absences,variable_performance]].groupby(variable_performance).describe().T
```

#Nous analysons deuc varibales donc une qualitative(PerformanceScore) et autre quantitative(Absences)

```python
df.groupby("PerformanceScore")["Absences"].agg(
    moyenne = "mean",
    mediane = "median",
    variance = "var",
    ecart_type = "std",
    minimum = "min",
    Q1 = lambda x: x.quantile(0.25),
    Q2 = "median",
    Q3 = lambda x: x.quantile(0.75),
    maximum = "max",
    IQR = lambda x: x.quantile(0.75) - x.quantile(0.25),
    etendue = lambda x: x.max() - x.min()
)
```
