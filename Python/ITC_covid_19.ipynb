##  Tests : Chi², Fisher, V de Cramér

```python
chi2, p_chi2, dof, expected = stats.chi2_contingency(table_contegence)
print(f"Test du Chi² : chi2 = {chi2:.3f}, p = {p_chi2:.3e}, ddl = {dof}")
```

```python
from scipy.stats import fisher_exact

if table_contegence.shape == (2, 2):
    oddsratio, p_fisher = fisher_exact(table_contegence)
    print(f"Test exact de Fisher : odds ratio = {oddsratio:.3f}, p = {p_fisher:.3e}")
else:
    print("La table_contegence n'est pas 2x2 : Fisher exact non applicable directement.")
```

```python

# V de Cramér
n = table_contegence.to_numpy().sum()
phi2 = chi2 / n
r, k = table_contegence.shape
phi2_corr = max(0, phi2 - (k-1)*(r-1)/(n-1))
r_corr = r - (r-1)**2/(n-1)
k_corr = k - (k-1)**2/(n-1)
cramers_v = np.sqrt(phi2_corr / min((k_corr-1), (r_corr-1)))
print(f"V de Cramér : {cramers_v:.3f}")
```

### Conclusion

- Si p(Chi²) > 0.05 → on accepte  H0 : les variables ne sont pas associées.  
- V de Cramér mesure la **force** de l'association (proche de 0 → faible, proche de 1 → forte).  
- Toujours interpréter en regardant les proportions (% par sexe) et le contexte métier.
