```python
score_passé = df["Previous_Scores"]
score_actuel= df["Exam_Score"]
# --- Covariance ---
cov_xy = np.cov(score_actuel, score_actuel, ddof=1)[0, 1]

# --- Corrélation de Pearson ---
pearson_r, pearson_p = stats.pearsonr(score_passé, score_actuel)

# --- Corrélation de Spearman (monotone) ---
spearman_rho, spearman_p = stats.spearmanr(score_passé, score_actuel)

# --- Corrélation de Kendall (τ) ---
kendall_tau, kendall_p = stats.kendalltau(score_passé, score_actuel)

print(" Covariance :", cov_xy)
print(f" Corrélation de Pearson : r = {pearson_r:.3f} (p = {pearson_p:.3e})")
print(f" Corrélation de Spearman : ρ = {spearman_rho:.3f} (p = {spearman_p:.3e})")
print(f" Corrélation de Kendall : τ = {kendall_tau:.3f} (p = {kendall_p:.3e})")
```

**D'après notre test de correlation de Pearson,nous remarquons que la correlation est tres faible ou presque nulle entre ces deux variables donc il n'existre pas une relation entre les variabes.**

```python
var_sexe="Gender"
var_score="Exam_Score"

df[[var_score, var_sexe]].groupby(var_sexe).describe().T
```

```python
from scipy.stats import shapiro, levene, mannwhitneyu

groupes = [g[var_score].dropna().values for _, g in df.groupby(var_sexe)]
labels = df[var_sexe].unique()

for label, g in df.groupby(var_sexe):
    vals = g[var_score].dropna()
    stat, p = shapiro(vals.sample(min(len(vals), 500), random_state=42))
    print(f"Shapiro pour {label} : stat={stat:.3f}, p={p:.3f}, n={len(vals)}")

g1, g2 = groupes
stat_lev, p_lev = levene(g1, g2)
print(f"\nTest de Levene (égalité des variances) : stat={stat_lev:.3f}, p={p_lev:.3f}")
```

```python
t_stat, t_p = stats.ttest_ind(g1, g2, equal_var=True)  # Student
print("t-test indépendant ")
print(f"t = {t_stat:.3f}, p = {t_p:.3e}\n")


t_stat, t_p = stats.ttest_ind(g1, g2, equal_var=False)  # Welch
print("t-test indépendant (Welch)")
print(f"t = {t_stat:.3f}, p = {t_p:.3e}\n")

mw_stat, mw_p = mannwhitneyu(g1, g2, alternative="two-sided")
print("Test de Mann–Whitney")
print(f"U = {mw_stat:.3f}, p = {mw_p:.3e}")
```

###  Conclusion

1. Regarder la normalité (Shapiro + graphes)  
2. Regarder les variances (Levene)  
3. Utiliser t-test (Welch) si c’est acceptable, sinon Mann–Whitney  
4. Conclure en français simple sur :  
   - l’existence ou non d’une différence  
   - le sens et l’ampleur de la différence (moyennes + graphes)

## Mettons en place un test pour conclure

```python
chi2, p_chi2, dof, expected = stats.chi2_contingency(table_contegince)
print(f"Test du Chi² : chi2 = {chi2:.3f}, p = {p_chi2:.3e}, ddl = {dof}")
```

```python
from scipy.stats import fisher_exact

if table_contegince.shape == (2, 2):
    oddsratio, p_fisher = fisher_exact(table_contegince)
    print(f"Test exact de Fisher : odds ratio = {oddsratio:.3f}, p = {p_fisher:.3e}")
else:
    print("La table_contegince n'est pas 2x2 : Fisher exact non applicable directement.")
```

```python
# V de Cramér
n = table_contegince.to_numpy().sum()
phi2 = chi2 / n
r, k = table_contegince.shape
phi2_corr = max(0, phi2 - (k-1)*(r-1)/(n-1))
r_corr = r - (r-1)**2/(n-1)
k_corr = k - (k-1)**2/(n-1)
cramers_v = np.sqrt(phi2_corr / min((k_corr-1), (r_corr-1)))
print(f"V de Cramér : {cramers_v:.3f}")
```

### 3.3 Conclusion

- Si p(Chi²) < 0.05 → on rejette H0 : les variables sont associées.  
- V de Cramér mesure la **force** de l'association (proche de 0 → faible, proche de 1 → forte).  
- Toujours interpréter en regardant les proportions (% par sexe) et le contexte métier.
Dans notre cas nous remarquons que **Test du Chi² : chi2 = 1.400, p = 2.366e-01, ddl = 1** qui signifie il n'ya aucune associations entre ces le genre et l'accés à l'internet.
